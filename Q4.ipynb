```python
driver=webdriver.Chrome()

driver.get('https://www.flipkart.com/')
time.sleep(3)
```

```python
Name=[]
MRP=[]
Offer_Price=[]
Ram_Rom=[]
Camera=[]
Battery=[]
Processer=[]
Product_URL=[]

try:
    Name_tag=driver.find_elements(By.XPATH,'//div[@class="_4rR01T"]')
    for i in Name_tag[0:10]:
        Name.append(i.text)
        
    marp_tag=driver.find_elements(By.XPATH,'//div[@class="_3I9_wc _27UcVY"]')
    for i in marp_tag[0:10]:
        MRP.append(i.text)
        
    Offer_Price_tag=driver.find_elements(By.XPATH,'//div[@class="_30jeq3 _1_WHN1"]')
    
    for i in Offer_Price_tag[0:10]:
        Offer_Price.append(i.text)
        
    Ram_Rom_tag=driver.find_elements(By.XPATH,'//li[@class="rgWa7D"][1]')
    for i in Ram_Rom_tag[0:10]:
        Ram_Rom.append(i.text)
        
    Camera_tag=driver.find_elements(By.XPATH,'//li[@class="rgWa7D"][3]')
    for i in Camera_tag[0:10]:
        Camera.append(i.text)
        
    Battery_tag=driver.find_elements(By.XPATH,'//li[@class="rgWa7D"][4]')
    for i in Battery_tag[0:10]:
        Battery.append(i.text)
        
    Processer_tag=driver.find_elements(By.XPATH,'//li[@class="rgWa7D"][5]')
    for i in Processer_tag[0:10]:
        Processer.append(i.text)
        
    Product_URL_tag=driver.find_elements(By.XPATH,'//a[@class="_1fQZEK"]')
    for i in Product_URL_tag[0:10]:
        Product_URL.append(i.get_attribute('href'))
        
        
               
        
except Exception as e:
    Name.append('No Information')
    MRP.appenda('No MRP')
    Offer_Price('No Offer Price')
    Ram_Rom.append ('No Information')
    Camera.append('No Information')
    Battery.append('No Information')
    Processer.append('No Information')
    Product_URL.append('No information')
    
    
    
```
