#### MODEL BUILD 

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
from sklearn.linear_model import LogisticRegression
```

```python
maxAcc=0
maxRE=0
for i in range (1,200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=i)
    lg=LogisticRegression()
    lg.fit (x_train,y_train)
    lg_pred=lg.predict(x_test)
    acc=accuracy_score(lg_pred,y_test)
    if acc>maxAcc:
        maxAcc=acc
        maxRE=i

print ('Best Accuracy Score:',maxAcc,'Best Random Estate Score:',maxRE)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import GradientBoostingClassifier,AdaBoostClassifier,BaggingClassifier

from sklearn.model_selection import cross_val_score
from sklearn import metrics
```

```python
model_class={}
def classify_Model(model):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=36)
    model.fit(x_train,y_train)
    pred_model=model.predict (x_test)
    print ('Best Traning Model:', model.score(x_train,y_train))
    print('Best Accuracy Score:', accuracy_score(pred_model,y_test))
    print ('\n')
    
    score=cross_val_score(model,x,y,cv=9)
    print('Best CV SCore:', np.mean(score))
    
    fpr,tpr,thresold=metrics.roc_curve(pred_model,y_test)
    roc_auc=metrics.auc(fpr,tpr)
    display=metrics.RocCurveDisplay(fpr=fpr,tpr=tpr,roc_auc=roc_auc,estimator_name=model)
    display.plot()
    model_class[model]=round(np.mean(score)*100,2)   
```

```python
model_class
```

```python
model=LogisticRegression()
classify_Model(model)
```

```python
model=RandomForestClassifier()
classify_Model(model)
```

```python
model=ExtraTreesClassifier()
classify_Model(model)
```

```python
model=DecisionTreeClassifier()
classify_Model(model)
```

```python
model=GaussianNB()
classify_Model(model)
```

```python
model=GradientBoostingClassifier()
classify_Model(model)
```

```python
model=AdaBoostClassifier()
classify_Model(model)
```
