# Model Buildup

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.linear_model import LogisticRegression
```

```python
maxAcc=0
maxRE=0
for i in range (1,200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=i)
    lr=LogisticRegression()
    lr.fit (x_train,y_train)
    pred_lr=lr.predict (x_test)
    acc=accuracy_score(pred_lr,y_test)
    if acc>maxAcc:
        maxAcc=acc
        maxRE=i

print ("Best Accuracy Score:",maxAcc,"Best Random Estate:",maxRE)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier,AdaBoostClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix,roc_curve

from sklearn.model_selection import cross_val_score
```

```python
def classifyModel(model):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=143)
    model.fit (x_train,y_train)
    pred_model=model.predict (x_test)
    print ("Model Traning Score:", model.score(x_train,y_train))
    print ("Best Accuracy Score:",accuracy_score(pred_model,y_test))
    print ('\n')
    
    print ("Classification Report:",classification_report(pred_model,y_test))
    print ("Confusion Matrix:",confusion_matrix(pred_model,y_test))
    print ('\n')
    
    score=cross_val_score(model,x,y,cv=7)
    print ("CV Score:", np.mean(score))
    
    
```

```python
model=RandomForestClassifier()                                                         # Accuracy Score: 0.86
classifyModel(model)                                                                   # CV SCore: 0.82
                                                                                       # Model Traning Score: 0.99
```

```python
model=LogisticRegression()                                                             # Accuracy Score: 0.82
classifyModel(model)                                                                   # CV SCore: 0.79
                                                                                       # Model Traning Score: 0.80
```

```python
model=SVC()                                                                            # Accuracy Score: 0.83
classifyModel(model)                                                                   # CV SCore: 0.80
                                                                                       # Model Traning Score: 0.84
```

```python
model=GradientBoostingClassifier()                                                     # Accuracy Score: 0.86
classifyModel(model)                                                                   # CV SCore: 0.82
                                                                                       # Model Traning Score: 0.86
```

```python
model=AdaBoostClassifier()                                                             # Accuracy Score: 0.84
classifyModel(model)                                                                   # CV SCore: 0.79
                                                                                       # Model Traning Score: 0.83
```

```python
model=ExtraTreesClassifier()                                                           # Accuracy Score: 0.85
classifyModel(model)                                                                   # CV SCore: 0.81
                                                                                       # Model Traning Score: 1.0
```

```python
model=DecisionTreeClassifier()                                                         # Accuracy Score: 0.79
classifyModel(model)                                                                   # CV SCore: 0.76
                                                                                       # Model Traning Score: 1.0
```

After evaluating various classification models based on the difference in Accuracy score and cross-validation scores, RandomForestClassifier is identified as the most suitable model for the given dataset.
