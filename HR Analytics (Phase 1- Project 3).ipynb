# Model BuildUp

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
from sklearn.linear_model import LogisticRegression
```

```python
maxAcc=0
maxRE=0
for i in range (1,200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=i)
    lg=LogisticRegression()
    lg.fit (x_train,y_train)
    pred_lg=lg.predict(x_test)
    acc=accuracy_score(pred_lg,y_test)
    if acc>maxAcc:
        maxAcc=acc
        maxRE=i

print ("Best accuracy score:",maxAcc,"Best random Estate:",maxRE)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier,AdaBoostClassifier,BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

from sklearn.model_selection import cross_val_score
```

```python
def classifier_mod(model):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=138)
    model.fit (x_train,y_train)
    pred_model=model.predict(x_test)
    print ('Best Score Model:', model.score(x_train,y_train))
    print ('Best Accuracy Score:', accuracy_score(pred_model,y_test))
    print ('Best Confusion Matrix:', confusion_matrix(pred_model,y_test))
    print('\n')
    
    score=cross_val_score(model,x,y,cv=9)
    print('Besr CV Score:', np.mean(score))
    
    
    
    
    
    
```

```python
model=LogisticRegression()
classifier_mod(model)
```

```python
model= RandomForestClassifier()
classifier_mod(model)
```

```python
model= GaussianNB()
classifier_mod(model)
```

```python
model= ExtraTreesClassifier()
classifier_mod(model)
```

```python
model= DecisionTreeClassifier()
classifier_mod(model)
```

##### Observation
After evaluating various regression models based on Accuracy score and cross-validation scores, ExtraTreesClassifier is identified as the most suitable model for the given dataset.

```python
final_mod=ExtraTreesClassifier(criterion='gini',max_features='log2',n_estimators=100)
final_mod.fit (x_train,y_train)
final_mod_pred=final_mod.predict (x_test)
print ('Best Model Score:', final_mod.score (x_train,y_train))
print ('Best Accuracy Score:',accuracy_score(final_mod_pred,y_test))
sns.heatmap(confusion_matrix(final_mod_pred,y_test),annot=True)


score = cross_val_score(final_mod,x,y,cv=9)
print ('Best CV Score:', np.mean (score))
```
