# Model Buildup

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.ensemble import RandomForestClassifier
```

```python
maxAcc=0
maxRE=0
for i in range (1,200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=i)
    rc=RandomForestClassifier()
    rc.fit (x_train,y_train)
    pred_rc=rc.predict(x_test)
    acc=accuracy_score(pred_rc,y_test)
    if acc>maxAcc:
        maxAcc=acc
        maxRE=i
        
print ("Best Accuracy Score:",maxAcc,"Best Random Estate Score:",maxRE)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier,AdaBoostClassifier,BaggingClassifier
from sklearn.ensemble import RandomForestClassifier,ExtraTreesClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix,roc_curve

from sklearn.model_selection import cross_val_score
```

```python
def classificationModel(model):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=187)
    model.fit (x_train,y_train)
    print ('Model Traning Score:',model.score(x_train,y_train))
    pred_model=model.predict (x_test)
    print ('\n')
    
    print ("Best Accuracy Score:", accuracy_score(pred_model,y_test))
    print  ("Classification Report:",classification_report(pred_model,y_test))
    print ("Confusion Matrix:",confusion_matrix(pred_model,y_test))
    print ('\n')
    
    
    score=cross_val_score(model,x,y,cv=7)
    print ("CV Score:", np.mean(score))
    print ("Difference Between accuracy score and cv score:", accuracy_score(pred_model,y_test) - np.mean(score))
    
    
    
```

```python
model=LogisticRegression()                                                    #Accuracy Score: 0.89
classificationModel(model)                                                    # CV Score: 0.83
                                                                              # Model Traning Score : 0.84
```

```python
model=SVC()                                                                   #Accuracy Score: 0.90
classificationModel(model)                                                    # CV Score: 0.85
                                                                              # Model Traning Score : 0.88
```

```python
model=GradientBoostingClassifier()                                            #Accuracy Score: 0.95
classificationModel(model)                                                    # CV Score: 0.87
                                                                              # Model Traning Score : 1.0
```

```python
model=BaggingClassifier()                                                     #Accuracy Score: 0.94
classificationModel(model)                                                    # CV Score: 0.86
                                                                              # Model Traning Score : 0.99
```

```python
model=RandomForestClassifier()                                                #Accuracy Score: 0.96
classificationModel(model)                                                    # CV Score: 0.88
                                                                              # Model Traning Score : 1.0
```

```python
model=ExtraTreesClassifier()                                                  #Accuracy Score: 0.94
classificationModel(model)                                                    # CV Score: 0.88
                                                                              # Model Traning Score : 1.0
```

```python
model=DecisionTreeClassifier()                                                #Accuracy Score: 0.89
classificationModel(model)                                                    # CV Score: 0.83
                                                                              # Model Traning Score : 1.0
```

After evaluating various classification models based on the difference in Accuracy score and cross-validation scores, RandomForestClassifier is identified as the most suitable model for the given dataset.
