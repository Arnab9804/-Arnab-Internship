# Q1:

```python
driver=webdriver.Chrome()
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos#Top_videos')
```

```python
Rank=[]
Video_Name=[]
Artist=[]
Viewer=[]
date=[]

try:
    rank=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[1]')
    for i in rank:
        Rank.append(i.text)
except NoSuchElementException:
    Rank.append('No information')
try:
    name=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[2]')
    for i in name:
        Video_Name.append(i.text)
except NoSuchElementException:
    Video_Name.append('No information')
try:
    art=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[3]')
    for i in art:
        Artist.append(i.text)
except NoSuchElementException:
    Artist.append('No information')
    
try:
    view=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[4]')
    for i in view:
        Viewer.append(i.text)
except NoSuchElementException:
    Viewer.append('No information')
    
try:
    dat=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr/td[5]')
    for i in dat:
        date.append(i.text)
        
except NoSuchElementException:
    date.append ('No information')
    
        
    
```

```python
len(Rank),len(Video_Name),len(Artist),len(Viewer),len(date)
```

```python
import pandas as pd
df=pd.DataFrame({'Rank':Rank,'Video_Name':Video_Name,'Artist':Artist,'Viewer':Viewer,'date':date})
df
```

```python
driver=webdriver.Chrome()
driver.get('https://archive.ics.uci.edu/')
time.sleep(2)
```

```python
view_dataset_click=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div/div/div/a[1]').click()
```

```python
Dataset_URL=[]
start=0
end=3
for page in range(start,end):
    url=driver.find_elements(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"]')
    for i in url:
        Dataset_URL.append(i.get_attribute('href'))
    next_click=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[3]/div/button[2]')
    next_click.click()
    time.sleep(2)
    
        
```

```python
len(Dataset_URL)
```

```python
Datasset_name=[]
Data_type=[]
Task=[]
Attribute_type=[]
Feature_type=[]
Year=[]

for url in Dataset_URL:
    driver.get(url)
    time.sleep(2)
    
    try:
        name=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[1]/div[2]/div/h1')
        Datasset_name.append(name.text)
    except NoSuchElementException:
        Datasset_name.append('No information')
    try:
        d_type=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[1]/p')
        Data_type.append(d_type.text)
    except NoSuchElementException:
        Data_type.append('No information')
    try:
        task=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[3]/p')
        Task.append(task.text)
    except NoSuchElementException:
        Task.append('No information')
    try:
        attribute_type=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[2]/p')
        Attribute_type.append(attribute_type.text)
    except NoSuchElementException:
        Attribute_type.append('No infirmation')
    try:
        feature_type=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[4]/p')
        Feature_type.append(feature_type.text)
    except NoSuchElementException:
        Feature_type.append('No information')
    try:
        year=driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[1]/div[2]/h2')
        Year.append(year.text)
    except NoSuchElementException:
        Year.append('No information')
        
                
        
```

```python
len(Datasset_name),len(Data_type),len(Task),len(Attribute_type),len(Feature_type),len(Year)
```

```python
import pandas as pd
df=pd.DataFrame({'Datasset_name':Datasset_name,'Data_type':Data_type,'Task':Task,'Attribute_type':Attribute_type,'Feature_type':Feature_type,'Year':Year})
df
```

```python
driver.close()
```
