Q1:Write a python program to display all the header tags from wikipedia.org and make data frame.

```python
page=requests.get('https://en.wikipedia.org/wiki/Main_Page')

soup=BeautifulSoup(page.content)

Header=[]

for upper in soup.find_all('h2',class_='mp-h2'):
    Header.append(upper.text)
    
print (Header)
    
    
```

```python
import pandas as pd
```

```python
df=pd.DataFrame({'All Headers':Header})
df
```

Q7:Write a python program to scrape mentioned details from dineout.co.in and make data frame

```python
page=requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')


soup=BeautifulSoup(page.text)

Restaurant_name=[]

for i in soup.find_all('div',class_='restnt-info cursor'):
    Restaurant_name.append(i.text)
    
print(Restaurant_name)

```

```python
Place_name=[]

for i in soup.find_all('div',class_='restnt-loc ellipsis'):
    Place_name.append(i.text)
    
print(Place_name)
```

```python
Rating=[]

for i in soup.find_all('div',class_='restnt-rating rating-4'):
    Rating.append(i.text)
    
print (Rating)
    
```

```python
Images=[]

for  i in soup.find_all('img',class_='no-img'):
    Images.append(i['data-src'])
    
print (Images)
    
```

```python
import pandas as pd
```

```python


df=pd.DataFrame({'Resturent Name':Restaurant_name,'Place':Place_name,'Rating':Rating,'Images Link':Images})
df
```

Q6: Write a python program to scrape the details of most downloaded articles from AI in last 90 days.https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles Scrape below mentioned details and make data frame

```python
page=requests.get('https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles')
soup=BeautifulSoup(page.text)

paper_title=[]

for paper in soup.find_all('h2',class_='sc-1qrq3sd-1 gRGSUS sc-1nmom32-0 sc-1nmom32-1 btcbYu goSKRg'):
    paper_title.append(paper.text)
    
print(len(paper_title))
print(paper_title)
    
```

```python
Authors=[]

for i in soup.find_all('span',class_='sc-1w3fpd7-0 dnCnAO'):
    Authors.append(i.text)
    
print (len(Authors))
print (Authors)
    
```

```python
Published_Date=[]

for i in soup.find_all('span',class_='sc-1thf9ly-2 dvggWt'):
    Published_Date.append(i.text)
    
print (len(Published_Date))
print (Published_Date)
    
```

```python
URL=soup.find_all('li',class_='sc-9zxyh7-1 sc-9zxyh7-2 kOEIEO hvoVxs')





links=[]
Start_link='https://www.journals.elsevier.com'

for link in URL:
    page_URL=link.find('a')['href']
    links.append(Start_link+page_URL)
    
print(len(links))
print (links)

    
    

```

```python
df=pd.DataFrame({'Title':paper_title,'Paper_Links':URL,'Authors':Authors,'Published_Date':Published_Date})
df
```

Q5:Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and make data frame

```python
page=requests.get('https://www.cnbc.com/world/?region=world')

soup=BeautifulSoup(page.text)
data=soup.find('div',class_='LatestNews-headlineWrapper')

Headline=[]
Time=[]
news_Link=[]
start_link='https://www.cnbc.com'

for i in soup.find_all('a',class_='LatestNews-headline'):
    Headline.append(i.text)
print(len(Headline))

for i in soup.find_all('time',class_='LatestNews-timestamp'):
    Time.append(i.text)
print(len(Time))
```

```python
data=soup.find('ul',class_='LatestNews-list')

news_Link=[]
start_link='https://www.cnbc.com'
for link in data:
    page_link=link.find('a')['href']
    news_Link.append(start_link+page_link)
print(len(news_Link))

df2=pd.DataFrame({'Headline':Headline,'Time':Time,'News_Link':news_Link})
df2
    
    

```
